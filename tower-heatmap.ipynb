```python
colleges = []

import xml.etree.cElementTree as ET
# get an iterable
context = ET.iterparse('C:\Users\Maxim\Downloads\FO2013CollegeMap.kml', events=("start", "end"))
# turn it into an iterator
context = iter(context)

# get the root element
event, root = context.next()
prefix = '{http://earth.google.com/kml/2.2}'

'''
Format:

<Placemark>
    <name>Carnegie Mellon</name>
    <description><![CDATA[<div dir="ltr">Kaleb,<br>Steffen,<br>Melissa</div>]]></description>
    <styleUrl>#style36</styleUrl>
    <Point>
      <coordinates>-79.944809,40.442570,0.000000</coordinates>
    </Point>
  </Placemark>

'''

for event, elem in context:
    #print elem.tag
    if event == "end" and elem.tag == prefix+"Placemark":
        #pId = elem.find('name')
        
        metadata = elem.getchildren()
        name = metadata[0].text
        students = metadata[1].text.replace('<div dir="ltr">', '').replace('<br>', '').replace('</div>', '').split(',')
        coords = [float(i) for i in metadata[3].getchildren()[0].text.split(',')[:2]]
        
        print name, students, coords, len(students)
        colleges.append([name, students, coords])
        
        root.clear()
        continue
```

```python
print 'Total number of students represented: ', sum([len(i[1]) for i in colleges])
```
